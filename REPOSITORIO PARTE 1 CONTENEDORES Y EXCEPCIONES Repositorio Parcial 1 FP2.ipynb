## 1. Tipos de Contenedores (Colecciones)

En Python, los contenedores permiten agrupar múltiples valores. La principal diferencia entre ellos es su mutabilidad (si se pueden cambiar tras crearlos) y si mantienen un orden.
Listas (list): Son secuencias mutables y ordenadas. Se definen con corchetes [].

 - Ejemplo: mi_lista = [10, 20, 30]
 - Operación clave: mi_lista.append(40) añade un elemento al final.

Tuplas (tuple): Son secuencias inmutables y ordenadas. Se definen con paréntesis ().

 - Ejemplo: mi_tupla = (1, 2, 3)
 - Uso: Ideales para datos que no deben cambiar (como coordenadas o constantes).

Diccionarios (dict): Son contenedores asociativos que guardan parejas clave:valor. Se definen con llaves {}.

 - Ejemplo: mi_dicto = {"nombre": "Juan", "edad": 21}
 - Acceso: Se accede por la clave, no por posición: mi_dicto["nombre"].

Conjuntos (set): Colecciones no ordenadas y sin elementos repetidos. Útiles para álgebra de conjuntos (unión, intersección).

Guía detallada de los métodos esenciales:
### 1. Listas (list): Son mutables y permiten gestionar colecciones ordenadas de datos.

Añadir elementos:

* .append(valor): Añade un elemento al final de la lista.
* .insert(índice, valor): Inserta un elemento en una posición específica, desplazando los demás.
* .extend(secuencia): Añade todos los elementos de otra secuencia al final.

Eliminar elementos:

* .pop(índice): Elimina y devuelve el elemento en esa posición. Si no se indica índice, elimina el último.
* .remove(valor): Elimina la primera aparición del valor especificado.
del lista[índice]: Elimina físicamente el elemento en esa posición.

Otros:

* .index(valor): Devuelve el índice de la primera aparición del valor (da error si no existe).
* len(lista): Función que devuelve el número total de elementos.

### 2. Diccionarios (dict): Almacenan parejas de clave:valor. Las claves deben ser inmutables (como strings o números).

Acceso y Gestión:

* .get(clave, [valor_por_defecto]): Recupera el valor. Si la clave no existe, devuelve None (o el valor por defecto) en lugar de dar error.
* clave in diccionario: Operador para comprobar si una clave existe antes de acceder.

Modificación y Borrado:

* .pop(clave): Elimina la pareja y devuelve el valor asociado.
* .clear(): Vacía completamente el diccionario.
* del diccionario[clave]: Elimina la pareja asociada a esa clave.

Vistas de datos (útiles para bucles):

* .keys(): Devuelve una lista con todas las claves.
* .values(): Devuelve una lista con todos los valores.
* .items(): Devuelve parejas (clave, valor), ideal para usar en for clave, valor in dicto.items():.

### 3. Conjuntos (set): Colecciones no ordenadas de elementos únicos.

Modificación:

* .add(valor): Añade un elemento (si ya existe, no hace nada).
* .update(secuencia): Añade múltiples elementos desde otro contenedor.

Eliminación:

* .remove(valor): Elimina el valor. Si no existe, lanza un error (KeyError).
* .discard(valor): Elimina el valor. Si no existe, no hace nada (más seguro que remove).
* .pop(): Elimina y devuelve un elemento al azar.

### 4. Tuplas (tuple): Son inmutables, por lo que no tienen métodos para añadir o eliminar elementos una vez creadas.

* .count(valor): Cuenta cuántas veces aparece un valor.
* .index(valor): Busca un valor y devuelve su posición.

## 2. Bucles for y Recorrido de Matrices
Uso de range() y len()
- len(secuencia): Devuelve el número de elementos.
- range(stop) o range(start, stop, step): Genera una secuencia de números para iterar.

Recorrido de una Matriz (Lista de Listas)
- Una matriz se recorre habitualmente con bucles anidados. El primer bucle recorre las filas y el segundo las columnas.

````python
matriz = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

# Recorrido por índices (útil si necesitas modificar la matriz)
for i in range(len(matriz)):          # Recorre filas (i = 0, 1, 2)
    for j in range(len(matriz[i])):   # Recorre columnas de esa fila
        print(f"Elemento en [{i}][{j}]: {matriz[i][j]}")
````

### 2.2 Herramientas Avanzadas de Iteración

- A. enumerate(): Índice y Valor a la vez

Cuando necesitas saber en qué posición estás y qué valor hay en esa posición al mismo tiempo. Evita tener que usar el clásico contador i = 0 y luego i += 1.

````python
nombres = ["Ana", "Berto", "Clara"]

# En lugar de: for i in range(len(nombres)):
for indice, nombre in enumerate(nombres):
    print(f"Posición {indice}: {nombre}")

# Truco de examen: Puedes empezar el conteo en 1
for n, nombre in enumerate(nombres, start=1):
    print(f"Estudiante nº{n}: {nombre}")
````

- B. zip(): Iterar varios contenedores en paralelo

Muy útil si tienes, por ejemplo, una lista de "Nombres" y otra de "Notas" y quieres emparejarlos en un solo bucle.

````python
estudiantes = ["Luis", "Marta", "Elena"]
notas = [8, 9, 7]

for nombre, nota in zip(estudiantes, notas):
    print(f"{nombre} ha sacado un {nota}")
    
# Si las listas tienen distinto tamaño, zip se detiene en la más corta.
````

- C. reversed(): Recorrer hacia atrás

A veces en los exámenes piden recorrer una lista desde el final al principio. En lugar de hacer cálculos raros con range(), usa esto:

````python
for item in reversed(nombres):
    print(item) # Clara, Berto, Ana
````

- D. items() para Diccionarios

Como mencionamos antes, es la forma estándar de recorrer un diccionario obteniendo la clave y el valor simultáneamente.

````python
precios = {"Manzana": 1.5, "Pera": 2.0}

for fruta, precio in precios.items():
    print(f"La {fruta} cuesta {precio}€")
````

### 2.3 Aplicación en Matrices (El "Combo" Ganador)

Ejemplo: "Busca el elemento mayor de una matriz y dime su fila y columna". Usar enumerate anidado es la forma más elegante:

````python
matriz = [
    [10, 20],
    [30, 40]
]

for i, fila in enumerate(matriz):
    for j, valor in enumerate(fila):
        if valor > 25:
            print(f"Encontrado {valor} en fila {i}, columna {j}")
````

## 3. Excepciones
### 1. ¿Qué es una Excepción?

En Python, una excepción es un objeto que representa un error que ocurre durante la ejecución de un programa. Cuando el intérprete encuentra una situación que no puede manejar (como dividir por cero o leer un archivo que no existe), "lanza" una excepción y, si no se controla, el programa finaliza abruptamente.

### 2. Tratamiento de Excepciones (try / except)

Para evitar que el programa se detenga, se utilizan los bloques try y except.
- Bloque try: Contiene el código "sospechoso" que podría fallar.
- Bloque except: Contiene el código que se ejecutará solo si ocurre el error.

````python
try:
    a = int(input('a: '))
    b = int(input('b: '))
    print(a // b) # (1)
except ZeroDivisionError:
    print('No se puede dividir por cero') # (2)

# Flujo: Si ocurre el error en (1), 
# se saltan todas las líneas restantes del try y 
# se ejecuta inmediatamente el except en (2).
````

##### Control de múltiples excepciones

Un solo try puede tener varios except para manejar distintos tipos de fallos de forma específica:
* except ValueError: Para errores de conversión (ej. introducir letras cuando se espera un número).
* except ZeroDivisionError: Para divisiones por cero.
* Except genérico: Se usa except: al final para capturar cualquier otra excepción no prevista.
* Múltiples en uno: Se pueden agrupar excepciones en una tupla si el tratamiento es el mismo: except (ValueError, ZeroDivisionError):.

### 3. Bloques Adicionales: else y finally

Para un control total, Python permite añadir dos bloques más:
* else: Se ejecuta solo si NO hubo ninguna excepción en el bloque try. Es el lugar ideal para código que no necesita protección pero depende del éxito del try.
* finally: Se ejecuta siempre, ocurra o no una excepción. Su uso principal es liberar recursos, como cerrar archivos o conexiones de red.

### 4. Información de la Excepción

Las excepciones guardan detalles sobre el error. Puedes acceder a ellos usando la palabra clave as:Pythonexcept Exception as err:
```py
    print(f"Error detectado: {err}") # Muestra el mensaje del error
    print(type(err)) # Muestra el tipo de excepción (ej. <class 'ValueError'>)
```
### 5. Lanzamiento de Excepciones (raise)

Sintaxis: raise TipoDeError("Mensaje personalizado").
Propagación: Si usas raise solo (sin argumentos) dentro de un except, relanzas la excepción capturada para que otro bloque superior la gestione.

### 6. Asertos (assert)
Un aserto es una condición que debe ser cierta para que el programa continúe.
- Sintaxis: assert condición, "Mensaje de error".
- Si la condición es falsa, Python lanza automáticamente un AssertionError. Se usa principalmente durante el desarrollo para verificar que el estado del programa es el correcto.

### Resumen

* Tratamiento: try (intenta), except (si falla), else (si funciona), finally (siempre)
* Lanzamiento: raise para informar de errores propios.
* Validación: assert para condiciones obligatorias.

El Reto: "Gestor de Notas con Errores"

Imagina que tienes una lista de nombres de alumnos y una lista de sus notas. Tu objetivo es crear un diccionario que los relacione, pero los datos vienen "sucios" (pueden faltar notas o haber valores no numéricos).

Instrucciones:

Tienes estas dos listas:

    alumnos = ["Ana", "Berto", "Clara", "David"]

    notas_sucias = ["8.5", "no_presentado", "7.0", "9.2"]

Recorre la lista de alumnos usando enumerate.

Intenta convertir cada nota a un número decimal (float).

Manejo de Excepciones:

Si la conversión falla (porque es un texto), captura la excepción ValueError.

Muestra un mensaje que diga: "Error en el alumno [índice]: No se pudo procesar la nota de [nombre]".

Contenedores: Solo si la nota es válida, añádela a un diccionario llamado registro donde la clave sea el nombre del alumno y el valor sea su nota.


Bloque Final: Al terminar el bucle, usa un bloque finally o simplemente un print final para mostrar cuántos alumnos se procesaron con éxito.

Atención: Asi se asocian dos listas: Aunque no es relevante en este ejercicio si hay que tener en cuenta
```py
for alumno, nota in zip(alumnos, notas_sucias):
    print(alumno, nota)

registro = dict(zip(alumnos, notas_sucias))
print(registro)
```

````python
def gestor_notas(alumnos, notas_sucias):
    registro = {}
    for i, alumno in enumerate(alumnos):
        try:
            nueva_nota = float(notas_sucias[i])
        except ValueError as err:
            print(f"Error en el alumno {i}: No se pudo procesar la nota de {alumno}, Motivo: {err}")
        else:
            registro[alumno] = nueva_nota
        
    return registro

if __name__ == "__main__":
    alumnos = ["Ana", "Berto", "Clara", "David"]
    notas_sucias = ["8.5", "no_presentado", "7.0", "9.2"]
    resultado = gestor_notas(alumnos, notas_sucias)
    print(f"Registro final: {resultado}")
````

Nivel 1: El "Catcher" de tipos (Básico)

Enunciado: Crea una función llamada sumador_inteligente(a, b) que intente sumar dos parámetros.Si se puede sumar, muestra el resultado en el bloque else.Si falla porque los tipos no son compatibles (por ejemplo, sumar un número y una letra), captura la excepción TypeError y muestra un mensaje explicativo.Independientemente de lo que pase, imprime "Cálculo finalizado" en el bloque finally.

````python
def sumador_inteligente(a, b):
    try:
        resultado = a+b
    except TypeError:
        print(f"Error de Cálculo, no se puede sumar")
        return None
    else:
        print(f"Resultado: {resultado}")
        return resultado
    finally:
        print(f"Cálculo finalizado")

a = 10
b = 2.5
print(sumador_inteligente(a, b))
````

Nivel 2: Validación de datos con raise (Intermedio)

Enunciado: Escribe una función verificar_edad(edad) que reciba un número.Si el número es negativo, debes lanzar tú mismo una excepción de tipo ValueError con el mensaje: "La edad no puede ser negativa" usando raise.Si el número es mayor de 120, lanza un ValueError que diga: "Edad fuera de rango humano".Fuera de la función, llama a verificar_edad dentro de un bloque try/except para capturar esos errores y mostrar el mensaje del error usando as err.

````python
def verificar_edad(edad):
    if edad < 0:
        raise ValueError("La edad no puede ser negativa")
    elif edad > 120:
        raise ValueError("Edad fuera de rango humano")
    return f"Edad {edad} válida"

try:
    print(verificar_edad(15))
except ValueError as err:
    print(f"No se ha podido verificar edad por error: {err}")
````

Nivel 3: El buscador de Diccionarios (Avanzado / Examen)

Enunciado: Este ejercicio simula una base de datos de una universidad. Tienes este diccionario:

db = {"progra_1": "Aprobado", "progra_2": "Pendiente"}

Crea un programa que:

1. Pida al usuario el nombre de una asignatura por teclado.
2. Intente buscar esa asignatura en el diccionario db.
3. Manejo de Excepciones:
    * Si la asignatura no existe, Python lanzará un KeyError. Captúralo y dile al usuario: "Asignatura no matriculada".
    * Si el usuario pulsa Ctrl+C (interrupción de teclado), Python lanza KeyboardInterrupt. Captúralo y despídete amablemente: "Saliendo del programa...".
4. Usa un bloque else para imprimir la nota si todo fue bien.

````python
def buscador(asignatura, db):
    try:
        valor = db[asignatura]
    except KeyError:
        return "Asignatura no matriculada"
    except KeyboardInterrupt:
        return "Saliendo del programa..."
    else:
        return f"Estado de {asignatura}: {valor}"

db = {"progra_1": "Aprobado", "progra_2": "Pendiente"}

try:
    asignatura = input("Inserte asignatura: ")
    print(buscador(asignatura, db))
except KeyboardInterrupt:
    print("\nSaliendo del programa...")
````

Ejercicio 4: El flujo completo(try-except-else-finally)

Enunciado: Crea una función llamada divisor_seguro() que haga lo siguiente:Pida dos números enteros por teclado (input).Intente realizar la división entera (//) del primero entre el segundo. Tratamiento:Si el usuario introduce letras, captura el ValueError. Si el divisor es cero, captura el ZeroDivisionError. Éxito: Si no hay errores, imprime el resultado en el bloque else. Cierre: Independientemente de lo ocurrido, imprime "Fin de la operación" en el bloque finally.

````python
def divisor_seguro():
    try:
        n1 = int(input("Introduzca num 1: "))
        n2 = int(input("Introduzca num 2: "))
        resultado = n1 // n2
    except ValueError:
        print("Error: No se pueden introducir letras, solo números enteros.")
    except ZeroDivisionError:
        print("Error: El divisor no puede ser cero.")
    else:
        print(f"El resultado es: {resultado}")
    finally:
        print("Fin de la operación")

divisor_seguro()
````

Ejercicio 5: Lanzamiento y Propagación (raise)

Enunciado:
1.  Crea una función calcular_area_cuadrado(lado).
2.  Si el lado es menor que 0, lanza una excepción ValueError con el mensaje: "El lado no puede ser negativo". 
3.  Si el lado no es un número (ni int ni float), lanza un TypeError que diga: "El lado debe ser numérico". 
4.  Fuera de la función, crea un bucle que pida el lado al usuario.
5.  Usa un bloque try-except para capturar ambos errores y mostrar el mensaje personalizado (as err). 
6.  Si el dato es correcto, muestra el área y rompe el bucle (break).

````python
def calcular_area_cuadrado(lado):
    if lado < 0:
        raise ValueError("El lado no puede ser negativo")
    if type(lado) != int or type(lado) != float:
        raise TypeError("El lado debe ser númerico")
    
    return lado ** 2
    
while True:
    try:
        lado = input("Ingrese Lado: ")

        if lado.lower() == 'salir':
            break
        
        valor_lado = float(lado)
    
        area = calcular_area_cuadrado(valor_lado)
    
    except ValueError as err:
        print(f"Error de valor: {err}")
    except TypeError as err:
        print(f"Error de tipo: {err}")
    except KeyboardInterrupt:
        print("\nPrograma interrumpido por el usuario.")
        break
    else:
        print(f"El área del cuadrado con lado {valor_lado} es: {area}")
        break
    finally:
        print("--- Intento procesado ---")
````
