## 1. Clases y Objetos (Instanciación)

- Clase: Es la entidad abstracta o concepto (ej. el concepto de "Silla" o "Rectángulo"). Se define con la palabra class.
- Objeto (Instancia): Es la materialización concreta de esa clase. Se crea usando el nombre de la clase seguido de paréntesis: mi_objeto = MiClase().

## 2. El método __ init __ y los Atributos

Para que todos los objetos de una clase tengan las mismas características básicas al nacer, usamos el método mágico __ init __.

- Atributos de datos: Son variables que representan el estado del objeto (ej. la base y la altura de un rectángulo).
- Inicialización: Dentro de __ init __, asignamos valores a los atributos usando la notación self.nombre_atributo = valor.

````python
class Rectangulo:
    def __init__(self, base, altura):
        self.base = base      # Atributo de instancia
        self.altura = altura  # Atributo de instancia
````

## 3. El parámetro self

- self representa al objeto sobre el que se está actuando en ese momento.
- Python lo pasa implícitamente. Cuando tú haces r1.area(), Python realmente ejecuta algo parecido a Rectangulo.area(r1).
- Es obligatorio ponerlo como primer parámetro en todos los métodos de instancia.

## 4. Métodos de instancia (Comportamiento)

Los métodos son funciones definidas dentro de la clase que manipulan los datos del objeto.
- Se invocan con la notación de punto: objeto.metodo().
- Si un método quiere llamar a otro método del mismo objeto, debe usar self.otro_metodo().

````python
class Rectangulo:
    def __init__(self, base, altura):
        self.base = base
        self.altura = altura

    def area(self):
        # Usamos self para acceder a los datos de ESTE objeto concreto
        return self.base * self.altura
````

## Encapsulamiento II

## 1. El Concepto: ¿Por qué @property?

En otros lenguajes (como Java), se usan obligatoriamente los métodos getter y setter. En Python, queremos que el código sea limpio. Las __Propiedades__ permiten que un método se comporte externamente como un atributo normal, pero con control interno.

## 2. La Estructura

A. Atributo Privado
- Definimos el dato con doble guion bajo en el __ init __.

self.__precio = 0

B. El Decorador @property (El Getter)

- Sirve para leer el valor. No lleva paréntesis al usarlo.

````python
@property
def precio(self):
    return self.__precio
````

C. El Decorador @nombre.setter (El Setter)

- Sirve para modificar y validar el valor. Se activa cuando usamos el signo =.

````python
@precio.setter
def precio(self, nuevo_valor):
    if nuevo_valor < 0:
        raise ValueError("El precio no puede ser negativo")
    self.__precio = nuevo_valor
````

## 3. Ventaja en el Examen: La Interfaz Limpia

La diferencia clave es cómo interactúa el usuario con tu clase:

- Sin Propiedades (Semana 3): obj.set_precio(10) / print(obj.get_precio())
- Con Propiedades (Semana 4): obj.precio = 10 / print(obj.precio)
- __IMPORTANTE__: Externamente parece que accedes a la variable directamente, pero internamente Python está ejecutando tus métodos y validaciones.

__Contexto: Sistema de Gestión de una Tienda__

1. Clase Producto:
- Atributos privados: __nombre, __precio.
- Propiedad precio con setter: No puede ser negativo (lanza ValueError).
- Método __ str __: Que muestre el nombre y el precio con 2 decimales.
2. Lógica Externa:
- Crea una lista de 3 productos pidiendo los datos por teclado.
- Usa un bloque try-except dentro de un bucle para que, si el usuario mete un precio negativo o una letra, el programa no se corte y le pida el producto de nuevo.
- Al final, recorre la lista con un for usando enumerate y muestra el precio total de la compra.

````python
class Producto:
    def __init__(self, nombre, precio):
        self.precio = precio
        self.nombre = nombre
    @property
    def nombre(self):
        return self.__nombre
    @nombre.setter
    def nombre(self, valor):
        self.__nombre = valor # sino en init deberia de ir self.__precio
    @property
    def precio(self): return self.__precio
    @precio.setter
    def precio(self, valor):
        if valor < 0: raise ValueError
        self.__precio = valor
    def __str__(self):
        return f"{self.nombre}: {self.precio:.2f}"

# --- LÓGICA EXTERNA ---
lista_productos = []
print("--- Registro de Productos ---")

while len(lista_productos) < 3:
    try:
        nom = input(f"Nombre del producto {len(lista_productos) + 1}: ")
        pre = float(input(f"Precio de {nom}: "))
        
        # Creamos el objeto. Si el precio es < 0, saltará al except
        nuevo_prod = Producto(nom, pre)
        lista_productos.append(nuevo_prod)
        
    except ValueError as err:
        # Esto captura tanto si el usuario mete letras como si el precio es negativo
        print(f"Dato inválido: {err}. Inténtalo de nuevo.")
    except KeyboardInterrupt:
        print("\nSaliendo...")
        break

# --- MOSTRAR RESULTADOS (Recorrido de la lista) ---
print("\n--- Carrito de la Compra ---")
total = 0

# Usamos enumerate para practicar (Semana 2)
for i, prod in enumerate(lista_productos, start=1):
    print(f"{i}. {prod}") # Esto llama al __str__ automáticamente
    total += prod.precio # Accedemos a la @property precio

print("-" * 20)
print(f"TOTAL: {total:.2f}€")
````

## Variables de Clase (Atributos de Clase)

A diferencia de los atributos de instancia (que se crean con self y son distintos para cada objeto), las variables de clase se definen fuera del __init__ y son compartidas por todas las instancias.
- ¿Para qué sirven? Para guardar datos comunes (como un contador de objetos creados o un IVA fijo).
- Acceso: Se recomienda acceder a ellas a través del nombre de la clase (Clase.variable) o de self.__class__.variable.

````python
class Producto:
    contador_productos = 0  # Variable de clase (compartida)

    def __init__(self, nombre):
        self.nombre = nombre  # Atributo de instancia (único)
        # Cada vez que creamos uno, aumentamos el contador global
        Producto.contador_productos += 1
````

## Métodos de Clase (@classmethod)
Son métodos que no actúan sobre un objeto específico, sino sobre la clase en sí.
- Decorador: Se usa @classmethod.
- Parámetro cls: En lugar de self, reciben cls (que hace referencia a la clase).
- Uso común: Crear "constructores alternativos" (por ejemplo, crear un objeto a partir de un string o un diccionario).

````python
class Producto:
    contador_productos = 0

    def __init__(self, nombre, precio):
        self.nombre = nombre
        self.precio = precio
        Producto.contador_productos += 1

    @classmethod
    def cuantas_instancias(cls):
        # Usamos cls para acceder a la variable de clase
        return f"Se han creado {cls.contador_productos} productos."

    @classmethod
    def crear_anonimo(cls, precio):
        # Ejemplo de constructor alternativo
        return cls("Sin nombre", precio) 
````

| Característica | Atributo/Mét. de Instancia | Variable/Método de Clase      |
|----------------|-----------------------------|-------------------------------|
| Identificador  | self                        | cls / NombreClase             |
| Pertenece a    | Al objeto individual        | A la clase (común a todos)    |
| Se define      | Dentro de métodos con self  | Fuera de los métodos          |
| Decorador      | Ninguno                     | @classmethod                  |

Se está desarrollando una clase llamada Intervalo para representar intervalos cerrados de números reales del tipo [a, b].

El constructor ya se encuentra escrito (ver código suministrado), de forma que recibe dos parámetros (a y b) y los almacena como atributos del objeto. Se presupone que se pretende trabajar con intervalos cerrados y que los extremos pueden ser enteros o reales.

Añada a la definición de la clase Intervalo los siguientes métodos:

- __extremos__, que devuelve una tupla (a, b) con los extremos del intervalo.
- __longitud__, que devuelve la longitud del intervalo (esto es, b - a).
- __contiene__, que, recibiendo un número x, devuelve True si x está dentro del intervalo cerrado, y False en caso contrario.
- __desplaza__, que, recibiendo un número dx, desplaza el intervalo sumando dx a ambos extremos (modifica el propio objeto).
- __interseccion__, que, recibiendo otro objeto Intervalo, devuelve un nuevo objeto Intervalo que representa la intersección entre ambos intervalos.
- __get_str__, que devuelve una ristra con el formato exacto "[a, b]" (con una coma y un espacio). Se usará la conversión estándar a string de Python para a y b.

Además:

- El constructor deberá lanzar ValueError si a > b.
- El constructor deberá lanzar TypeError si a o b no son de tipo int o float.
- contiene deberá lanzar TypeError si x no es int o float.
- desplaza deberá lanzar TypeError si dx no es int o float.
- interseccion deberá lanzar TypeError si el parámetro no es un objeto Intervalo.
- interseccion deberá lanzar ValueError si los intervalos no se solapan (es decir, si la intersección es vacía).

Ejemplo:

para Intervalo(1, 5), longitud() devuelve 4, contiene(5) devuelve True, y get_str() devuelve "[1, 5]".

Si hacemos desplaza(-2), ahora el intervalo pasa a ser [-1, 3]. Si tenemos además Intervalo(2, 10), la intersección con [-1, 3] sería [2, 3].

````python
class Intervalo:
    """Intervalo cerrado [a, b] con control de errores."""

    def __init__(self, a, b):
        """Crea el intervalo con extremos a y b (cerrado)."""
        # NO MODIFIQUE EL CONSTRUCTOR
        # (salvo para completar controles si fuese necesario).
        if a > b:
            raise ValueError
        if not isinstance(a, (float, int)) or not isinstance(b, (float, int)):
            raise TypeError
        self.__a = a
        self.__b = b

    def extremos(self):
        return (self.__a, self.__b)

    def longitud(self):
        return self.__b - self.__a

    def contiene(self, x):
        if not isinstance(x, (int, float)):
            raise TypeError
        return self.__a <= x <= self.__b
        # return true or false si cumple

    def desplaza(self, dx):
        if not isinstance(dx, (int, float)):
            raise TypeError
        self.__a += dx
        self.__b += dx

    def interseccion(self, objeto):
        if not isinstance(objeto, Intervalo):
            raise TypeError
        # Sugerencia: Name Mangling
        """
        inicio_otro, fin_otro = objeto.extremos()
        nuevo_inicio = max(self.__a, inicio_otro)
        """
        nuevo_inicio = max(self.__a, objeto.__a)
        nuevo_fin = min(self.__b, objeto.__b)
        if nuevo_inicio > nuevo_fin:
            raise ValueError("Los intervalos no se solapan")
        return Intervalo(nuevo_inicio, nuevo_fin)

    def get_str(self):
        return f"[{self.__a}, {self.__b}]"
````

````python
class BadMatrix(Exception):
    """Exception for BadMatrix."""
    pass

def is_strictly_diagonally_dominant(matrix):
    """Return True if matrix is strictly diagonally dominant by rows."""
    n_filas = len(matrix)
    
    for i in range(n_filas):
        # 1. Comprobar si es cuadrada
        if len(matrix[i]) != n_filas:
            raise BadMatrix("Matriz no cuadrada")
        
        suma_otros = 0
        valor_diagonal = 0
        
        for j in range(n_filas):
            elemento = matrix[i][j]
            
            # 2. Control estricto de tipos: No se aceptan floats (como 0.1 o 5.0)
            if isinstance(elemento, float):
                raise BadMatrix("Dato no convertible a entero")
            
            # 3. Intento de conversión (para strings o tipos no válidos como None/listas)
            try:
                valor_int = int(elemento)
            except (ValueError, TypeError):
                raise BadMatrix("Dato no convertible a entero")
            
            # 4. Lógica de dominancia (usando valores absolutos)
            if i == j:
                valor_diagonal = abs(valor_int)
            else:
                suma_otros += abs(valor_int)
        
        # 5. Verificación de la fila: El de la diagonal DEBE ser mayor que la suma del resto
        if valor_diagonal <= suma_otros:
            return False
            
    # Si recorre todas las filas y ninguna falla, es dominante
    return True
````

Usando apropiadamente la clase Punto, modifica la clase Circulo suministrada para que disponga de las siguientes características:

1. Constructor que permita crear un círculo a partir de:

- radio (número int o float)
2. Si el radio recibido no es válido (radio ≤ 0), se debe lanzar la excepción BadCircle definida en el módulo solution.

- radio: devuelve/recibe un número. Si al asignar se recibe radio ≤ 0 se lanza BadCircle.
4. Propiedades de solo lectura:
- perimetro: 2·π·radio (usar math.pi)
- area: π·radio²
5. Método esta_contenido(p):
- Recibe un Punto p
- Devuelve True si el punto está dentro o en el borde del círculo, y False en caso contrario.
- Al comparar dos distancias resultantes de cálculos en coma flotante, para mitigar problemas de redondeo se recomienda compararlas al cuadrado.
6. Método interseca(otro):
- Recibe otro objeto Circulo
- Devuelve True si ambos círculos tienen al menos un punto en común (se cortan o son tangentes), y False si son disjuntos o uno está totalmente dentro del otro sin tocar.
- Condición geométrica:
    - Sean d la distancia entre centros; y r₁ y r₂ los respectivos radios:
    - Intersecan si | r₁ − r₂ | ≤ d ≤ r₁ + r₂.
7. Representación informal (str(circulo)):
- Debe devolver una cadena con el formato:
    - "<(x, y), r>" donde (x, y) es exactamente centro.get_str() y r es el radio tal cual (por ejemplo 2, 2.5, etc.)
- Ejemplo: si centro = Punto(-1, 2.5) y radio = 3 ⇒ "<(-1, 2.5), 3>"
8. Ejemplo de ejecución (orientativo)
- Círculo correcto: se imprime perímetro/área, contenido, intersección…
- Círculo erróneo (radio <= 0): se imprime Círculo erróneo

````python
import math
class Circulo:
    def __init__(self, centro, radio):
        self.centro = centro
        self.radio = radio
        # se validan con el setter
        if not isinstance(radio, (int, float)) or radio <= 0:
            raise BadCircle()

    @property
    def centro(self):
        return self.__centro

    @centro.setter
    def centro(self, valor):
        self.__centro = valor

    @property
    def radio(self):
        return self.__radio

    @radio.setter
    def radio(self, valor):
        if valor <= 0:
            raise BadCircle
        self.__radio = valor

    # propiedad que se usa como c.perimetro
    @property
    def perimetro(self):
        return 2*math.pi*self.radio

    # propiedad que se usa como c.area
    @property
    def area(self):
        return math.pi*self.radio**2

    def esta_contenido(self, p):
        suma = (p.x - self.centro.x)**2 + (p.y - self.centro.y)**2
        if suma <= self.radio**2:
            return True
        return False

    def interseca(self, otro):
        distancia = (self.centro.x - otro.centro.x)**2 + (self.centro.y - otro.centro.y)**2
        suma = (self.radio + otro.radio)**2
        diferencia = self.radio - otro.radio
        if abs(diferencia) <= distancia <= suma:
            return True
        return False

    def __str__(self):
        return f"<{self.centro.get_str()}, {self.radio}>"

# NO MODIFIQUE EL CODIGO DEBAJO DE ESTA LINEA


class Punto:
    """Punto con dos propiedades x e y."""

    def __init__(self, x, y):
        """Inicializa punto."""
        self.x = x
        self.y = y

    def get_str(self):
        """Devuelve punto como str."""
        return f"({self.x}, {self.y})"


class BadCircle(Exception):
    """Exception for BadCircle."""
    pass
````

# IMPORTANTE
1. ¿Qué pasa cuando usas self.radio?

Sí, exactamente. Cuando escribes self.radio (sin los guiones bajos), Python busca si existe una @property llamada radio.

* Si estás leyendo (print(self.radio)), se ejecuta el método bajo @property.
* Si estás asignando (self.radio = 5), se ejecuta el método bajo @radio.setter.

2. ¿Podrías usar self.__radio?

Poder, puedes. Al estar dentro de la misma clase Circulo, tienes acceso directo a la variable privada self.__radio.

PERO (y aquí está el truco): Si usas self.__radio, __te estás saltando a "los guardias de seguridad" (tus getters y setters).__

##### Escenario A: Usando el atributo privado (self.__radio)
```Python
def __init__(self, centro, radio):
    self.__radio = radio  # Acceso directo
```

Resultado: Si alguien crea un Circulo(centro, -50), el programa guardará el -50 sin rechistar. Tu validación del setter nunca se ejecutó porque fuiste directo al "cajón" privado.

##### Escenario B: Usando la propiedad (self.radio)
```Python
def __init__(self, centro, radio):
    self.radio = radio  # Esto activa el @radio.setter
```

- __Resultado__: Si alguien intenta pasar un -50, el setter se activa, ve que es menor que cero y lanza BadCircle. Tu objeto está protegido desde el segundo uno.

#### Dentro de tu clase, la recomendación es:

- Usa self.radio (la propiedad) siempre que quieras beneficiarte de las validaciones que escribiste en el setter.

- Usa self.__radio (el atributo privado) solo dentro de los propios métodos del getter y el setter, para evitar bucles infinitos.

Inicializador (constructor): Recibe el título, el autor y el año de publicación del libro; por este orden. Los libros al inicializarse no están prestados.

```py
libro1 = Libro('Cien años de soledad', 'Gabriel García Márquez', 1967)

libro.prestar()
print(libro1)  # "Cien años de soledad" por Gabriel García Márquez (1967) - Prestado
```

````python
# ESCRIBA AQUÍ LA CLASE
class Libro:
    # descomente la siguiente línea si quiere usarla
    __autor_número: dict[str, int] = {}  # diccionario autor: nº libros
    __total = 0  # Variable de clase que cuenta todos los libros que hay

    def __init__(self, titulo, autor, fecha):
        self.__titulo = titulo
        self.__autor = autor
        self.__fecha = fecha
        self.prestado = False
        # Inicia no prestado, uso el setter para inicializar

        # Con esto comprobamos si el autor está -> Suma 1
        # Y si no, se crea como Contador
        if autor in Libro.__autor_número:
            Libro.__autor_número[autor] += 1
        else:
            Libro.__autor_número[autor] = 1

        Libro.__total += 1

    @property
    def titulo(self):  # Solo lectura
        return self.__titulo

    @property
    def autor(self):  # Solo lectura
        return self.__autor

    @property
    def fecha(self):  # Solo lectura
        return self.__fecha

    @property
    def prestado(self):  # Lectura
        return self.__prestado

    @prestado.setter
    def prestado(self, valor):
        if not isinstance(valor, bool):
            raise ValueError("Valor booleano necesario")
        self.__prestado = valor

    def prestar(self):
        if self.prestado:  # Si es True
            raise ValueError("El libro ya está prestado")
        self.prestado = True

    def devolver(self):
        if not self.prestado:
            raise ValueError("El libro no estaba prestado")
        self.prestado = False

    @classmethod
    def total_libros(cls):
        return cls.__total

    @classmethod
    def publicados_autor(cls, autor):
        return cls.__autor_número.get(autor, 0)

    def __str__(self):
        estado = "Prestado" if self.prestado else "Disponible"
        return f"\"{self.titulo}\" por {self.autor} ({self.fecha}) - {estado}"
````

Clase CuentaBancaria 💳

* Inicializador (constructor): Recibe en una string el nombre del titular de la cuenta a crear. 
* Propiedades de instancia:
    * codigo_cuenta (solo lectura): string, código de la cuenta; inmutable, con el formato "CB-<em>nnn</em>", donde <em>nnn</em> son tres cifras numéricas que se deberán ir asignando a las sucesivas cuentas conforme vayan creándose, empezando por 001 y siguiendo con 002, 003, etc. (se recomienda usar el formato :03d en una f-string). Si no se resolviera la generación automática, utilícese el código de cuenta "CB-001". 
    * titular (solo lectura): string, nombre del titular de la cuenta.
    * saldo (solo lectura): flotante, saldo actual de la cuenta (inicialmente, 0.0).
* Métodos de instancia:
    * depositar(cantidad): Añade la cantidad al saldo, salvo si la cantidad fuera negativa, en cuyo caso se deberá lanzar una excepción ValueError.
    * retirar(cantidad): Resta la cantidad al saldo, salvo si el saldo fuera insuficiente o la cantidad fuera negativa, en cuyo caso se deberá lanzar una excepción ValueError.
* Métodos de clase:
    * obtener_total_cuentas(): Devuelve el número total de cuentas creadas.
    * total_saldo_titular(titular): Devuelve la suma de los saldos de todas las cuentas del titular.
* Representación informal:
    * Contendrá la información de la cuenta, siguiendo el formato:<código_cuenta> - Titular: <titular>, Saldo: <saldo>
    * Por ejemplo, para el siguiente código se debe mostrar lo indicado:
```py
cuenta1 = CuentaBancaria('José Rodríguez')
cuenta1.depositar(12)
print(cuenta1)  # CB-001 - Titular: José Rodríguez, Saldo: 12.0
cuenta2 = CuentaBancaria('Juan García')
cuenta2.depositar(33)
print(cuenta2)  # CB-002 - Titular: Juan García, Saldo: 33.0
```

````python
class CuentaBancaria:
    # descomente la siguiente línea si quiere usarla
    __titular_suma: dict[str, float] = {}  # diccionario titular: suma saldos
    __cuentas_total = 0

    def __init__(self, nombre_titular):
        CuentaBancaria.__cuentas_total += 1
        self.__titular = nombre_titular
        self.__saldo = 0.0  # Inicia a 0
        self.__codigo_cuenta = f"CB-{CuentaBancaria.__cuentas_total:03d}"

        if nombre_titular not in CuentaBancaria.__titular_suma:
            CuentaBancaria.__titular_suma[nombre_titular] = 0.0

    @property
    def codigo_cuenta(self):
        return self.__codigo_cuenta

    @property
    def titular(self):
        return self.__titular

    @property
    def saldo(self):
        return self.__saldo

    def depositar(self, cantidad):
        if cantidad < 0:
            raise ValueError("Cantidad negativa no permitida")
        self.__saldo += cantidad
        CuentaBancaria.__titular_suma[self.titular] += cantidad

    def retirar(self, cantidad):
        if cantidad < 0:
            raise ValueError("Cantidad negativa no permitida")
        if cantidad > self.__saldo:
            raise ValueError()
        self.__saldo -= cantidad
        CuentaBancaria.__titular_suma[self.titular] -= cantidad

    @classmethod
    def obtener_total_cuentas(cls):
        return cls.__cuentas_total

    @classmethod
    def total_saldo_titular(cls, titular):
        return cls.__titular_suma.get(titular, 0.0)

    def __str__(self):
        return f"{self.codigo_cuenta} - Titular: {self.titular}, Saldo: {self.saldo}"
````
