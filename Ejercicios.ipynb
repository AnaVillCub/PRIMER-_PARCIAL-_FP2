* Propiedad combustible:
    * Getter: Devuelve el nivel actual.
    * Setter: Si el valor es menor que 0 o mayor que 100, lanza un ValueError.
* Método conducir(self, km):
    * Resta 0.1 de combustible por cada km.
    * Si el combustible llega a 0, lanza un ValueError con el mensaje "Sin combustible". No permitas que el combustible sea negativo.
* Representación: [Matrícula] - Combustible: XX%.

```python
class Vehiculo:
    def __init__(self, matricula, combustible):
        self.__matricula = matricula
        self.combustible = combustible
        if not (0 < combustible < 100): raise ValueError()

    @property
    def matricula(self):
        return self.__matricula
    @property
    def combustible(self):
        return self.__combustible
    @combustible.setter
    def combustible(self, valor):
        if valor > 100 or valor < 0: raise ValueError()
        self.__combustible = valor

    def conducir(self, km):
        consumo = 0.1*km
        if self.combustible - consumo < 0:
            self.combustible = 0
            raise ValueError("Sin combustible")
        self.combustible -= consumo

    def __str__(self):
        return f"[{self.__matricula}] - Combustible: {self.combustible}%"
    
```
